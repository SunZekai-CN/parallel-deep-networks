# 16 Core Machine Results

```python
import seaborn as sns
import pandas as pd
```

```python
metrics = pd.read_csv('../../data/exp1/log/training_metrics.csv')
metrics.head()
```

## Training Time - 2 Processes vs 1 Process

```python
sns.catplot(x='threads', y='training_time', hue='is_distributed', kind='strip', col='architecture', data=metrics)
```

## Accuracy - Architecture Comparison

```python
sns.catplot(x='threads', y='accuracy', hue='is_distributed', kind='point', col='architecture', data=metrics)
```
